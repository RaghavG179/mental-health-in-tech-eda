## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ----------------------------------------
# üîß Data Wrangling
# ----------------------------------------

# Clean Age column
print("Before cleaning age column:", df['Age'].min(), "-", df['Age'].max())
df = df[(df['Age'] >= 14) & (df['Age'] <= 100)]
print("After cleaning age column:", df['Age'].min(), "-", df['Age'].max())

# Normalize Gender column
print("Unique gender values before cleaning:\n", df['Gender'].unique())
df['Gender'] = df['Gender'].astype(str).str.strip().str.title()
print("\nUnique gender values after cleaning:\n", df['Gender'].unique())

# Strip all object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
````

### What All Manipulations Have You Done and Insights You Found?

- Standardized the `Gender` column by fixing inconsistent formats and capitalizations (e.g., "male", "M", "Male-ish" ‚Üí "Male").
- Trimmed leading/trailing spaces in all object-type columns for consistency.
- Planned to drop non-analytical columns such as `Timestamp` and `comments` (if present).
